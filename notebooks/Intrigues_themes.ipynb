# Intrigues et thèmes
We import `pandas` and `pygal` and read in the csv-File. Printing out the table to check if it is ok: 

```python
import pandas as pd
import pygal
df = pd.read_csv("intrigues_themes.csv", sep=';')    
df
```

```python
import pygal
from pygal.style import Style


custom_style = Style(
    font_family='Roboto,Helvetica,Arial,sans-serif',
    label_font_size=12,
)
c = pygal.Bar(style=custom_style, y_title='Thèmes',
    width=1900,
    x_label_rotation=270,)
c.add('Intrigues sentimentales', df['a'])
#c.add('Intrigues galantes', df['b'])
#c.add('Episodes pornographiques', df['c'])
#c.add('Aventures romanesques', df['d'])
c.add('Voyages', df['e'])
#c.add('Romans d`éducation', df['m'])
c.x_labels = df['year']



c.render_to_file('themes.svg')
display(HTML(base_html.format(rendered_chart=c.render(is_unicode=True))))
```

```python
import pygal                                                       # First import pygal
bar_chart = pygal.Bar()                                            # Then create a bar graph object

bar_chart.add ('1752', [9, 4, 4, 6, 3 ,7 ,2 ,1, 0, 4, 4, 2, 1 , 1])# Add some values   	1752 
bar_chart.add('1796', [23, 6, 1, 11, 5, 12, 1, 1, 1, 3, 1, 3, 3, 4]) # Add some values
#bar_char.add(raw["a"])
bar_chart.title = 'Intrigues et thèmes' #add the Bar chart title
bar_chart.x_labels = 'a) Intrigues sentimentales et amours pastorales', 'b) Intrigues galantes et libertines.', 'c) Episodes scabreux, pornographiques', 'd) Aventures romanesques, mystères, etc.','e) Voyages', 'f) Peinture et critique des mœurs', 'g) Intrigues politiques et de cour; guerres', 'h) Eléments historiques et pseudo-historiques','i) Chevalerie, genre troubadour', 'j) Contes des fées, éléments merveilleux, légende, Bible', 'k) Allégorie', 'l) Thèmes et idées philosophiques, utopie', 'm) Ouvrages didactiques et pédagogiques, romans d`éducation', 'n) Intentions moralisatrices, histoires édifiantes, enseignement religieux'
#' #adding the x labels 
bar_chart.render_to_file('intrigues_et_themes.svg') # Save the svg to a file
display(HTML(base_html.format(rendered_chart=bar_chart.render(is_unicode=True))))
```
